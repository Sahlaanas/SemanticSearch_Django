```python
import re


def generate_dummy_data(num_samples=10):


  llm = GoogleGenerativeAI(model="gemini-pro")

  prompt_template = """
  Generate {num_samples} JSON objects, each containing a "question" and "answer" field.
  The questions should be related to common e-commerce topics and should be FAQ. The answers should be accurate and informative.

  **Example:**
  [
      {{"question": "What are your shipping options?", "answer": "We offer standard and express shipping."}},
      {{"question": "How do I return an item?", "answer": "Please contact our customer support to initiate a return."}}
  ]
  """

  prompt = PromptTemplate(
      input_variables=[],
      template=prompt_template
  )

  response = llm(prompt.format(num_samples=num_samples))
  print(response)

  # Extract JSON string using regex
  match = re.search(r"\[.*\]", response, re.DOTALL)
  if match:
      response = match.group(0)
  else:
      print("Error: Could not find JSON array in response.")
      return []

  try:
      data = json.loads(response)
      return data
  except json.JSONDecodeError:
      print("Error: Invalid JSON response from the LLM.")
      return []

if __name__ == "__main__":
  num_samples = 10  # Adjust as needed
  dataset = generate_dummy_data(num_samples)

  # Save the dataset to a JSON file
  with open("dummy_dataset.json", "a") as f:
      json.dump(dataset, f, indent=4)
```

```python
import pandas as pd
faq_df = pd.DataFrame({
    'question': test_questions,
    'label': similarity_list,
    'predicted_similarity': pred_similarity_list
})
```

```python
import numpy as np
```

```python
def calculate_mrr(faq_df):
    # Initialize a list to store the reciprocal ranks for each query
    reciprocal_ranks = []

    # Iterate over each unique question
    for question in faq_df['question'].unique():
        # Get all rows for this question
        question_df = faq_df[faq_df['question'] == question]

        # Sort the answers by predicted similarity in descending order
        question_df = question_df.sort_values(by='predicted_similarity', ascending=False)

        # Get the ground truth label (the correct answer for this question)
        correct_answer_idx = question_df['label'].idxmax()  # The answer with the highest cosine similarity label

        # Find the rank of the correct answer in the sorted list
        rank = question_df.index.get_loc(correct_answer_idx) + 1  # Rank starts at 1
        # Compute the reciprocal of the rank and append it to the list
        reciprocal_ranks.append(1 / rank)


    # Calculate Mean Reciprocal Rank (MRR)
    mrr = np.mean(reciprocal_ranks)
    return mrr

# Calculate and display the MRR
mrr_score = calculate_mrr(faq_df)
print(f"Mean Reciprocal Rank (MRR): {mrr_score:.4f}")
```

```python
# Function to compute DCG
def dcg(relevance_scores):
    return np.sum(relevance_scores / np.log2(np.arange(2, len(relevance_scores) + 2)))

# Function to compute NDCG
def ndcg(predicted_relevance, ground_truth_relevance, k=5):
    # Sort the predicted relevance scores and ground truth relevance in descending order
    predicted_relevance = np.array(predicted_relevance)[:k]
    ground_truth_relevance = np.array(ground_truth_relevance)[:k]

    # Compute DCG and IDCG
    dcg_score = dcg(predicted_relevance)
    idcg_score = dcg(np.sort(ground_truth_relevance)[::-1])  # Sort in descending order for ideal ranking

    # Calculate NDCG
    return dcg_score / idcg_score if idcg_score != 0 else 0

# Calculate NDCG
ndcg_score = ndcg(faq_df['predicted_similarity'], faq_df['label'], k=5)
print(f"NDCG@5: {ndcg_score:.4f}")
```
